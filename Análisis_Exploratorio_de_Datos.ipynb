```python
import pandas as pd
import requests
from zipfile import ZipFile
from io import BytesIO
```

```python
url = 'http://www.dgis.salud.gob.mx/descargas/datosabiertos/nacimientos/sinac_2023.zip?V=2024.05.14'

r = requests.get(url)
if r.status_code == 200:
    z = ZipFile(BytesIO(r.content))
    z.extractall()
z.extractall()
print(z.namelist())
```

```python
df = pd.read_csv('Nacimientos_2023.csv')
```

```python
df.columns
```

```python
df = df[['SEXO','TALLA','PESO','TIPOCESAREA','EDADGESTACIONAL','FECHACERTIFICADO', 'ENTIDADFEDERATIVAPARTO','EDAD','EDADPADRE']]

nuevos_nombres = {
    'SEXO': 'SEXO',
    'TALLA': 'TALLA (CM)',
    'PESO': 'PESO (GRAMOS)',
    'TIPOCESAREA': '¿CESARIA?',
    'EDADGESTACIONAL': 'EDAD GESTACIONAL (SEMANAS)',
    'FECHACERTIFICADO': 'FECHA DE NACIMIENTO',
    'ENTIDADFEDERATIVAPARTO': 'ENTIDAD FEDERATIVA',
    'EDAD': 'EDAD MADRE',
    'EDADPADRE': 'EDAD PADRE'
}

df['FECHACERTIFICADO'] = pd.to_datetime(df['FECHACERTIFICADO'], format="%d/%m/%Y")

df = df.rename(columns=nuevos_nombres)
```

```python
df_sonora = df[df['ENTIDAD FEDERATIVA'] == 26]
entidad_federativa = {26: 'Sonora'}
df_sonora['ENTIDAD FEDERATIVA'] = df_sonora['ENTIDAD FEDERATIVA'].replace(entidad_federativa)

df_sonora
```

```python
sexo = {1: 'Masculino',2: 'Femenino',0: 'No especificado',9:'No especificado'}
cesaria = {8:'No',0:'Si',1:'Si',2:'Si'}

df_sonora['SEXO'] = df_sonora['SEXO'].replace(sexo)
df_sonora['¿CESARIA?'] = df_sonora['¿CESARIA?'].replace(cesaria)

df_sonora
```

```python
print('De el data frame anterior podemos obtener los siguientes datos interesantes:')
print()
print('La edad de la mujer que tuvo hijos más joven en Sonora en 2023 tenía:', df_sonora['EDAD MADRE'].min())
print('La edad de la mujer que tuvo hijos más grande en Sonora en 2023 tenía:', df_sonora['EDAD MADRE'].max())
print('El promedio de edad de las mujeres que tuvieron hijos en Sonora en 2023 es de:',df_sonora['EDAD MADRE'].mean())
print()
print('La edad del hombre que tuvo hijos más joven en Sonora en 2023 tenía:',df_sonora['EDAD PADRE'].min())
print('La edad del hombre que tuvo hijos más grande en Sonora en 2023 tenía:',df_sonora['EDAD PADRE'].max())
print('El promedio de edad de los hombres que tuvieron hijos en Sonora en 2023 es de:',df_sonora['EDAD PADRE'].mean())
print()
print('En 2023 en Sonora, nacieron más niños de sexo:', df_sonora['SEXO'].value_counts().idxmax())
print('En 2023 en Sonora, ¿la mayoría de nacidos fue por cesaria?', df_sonora['¿CESARIA?'].value_counts().idxmax())
print()
print('El nacido mas pesado en Sonora en 2023 pesó:', df_sonora['PESO (GRAMOS)'].max())
print('El nacido mas liviano en Sonora en 2023 pesó:', df_sonora['PESO (GRAMOS)'].min())
print('El peso promedio de los nacidos en Sonora en 2023 es:', df_sonora['PESO (GRAMOS)'].mean())
print()
print('El nacido mas grande en Sonora en 2023 midió:', df_sonora['TALLA (CM)'].max())
print('El nacido mas chico en Sonora en 2023 midió:', df_sonora['TALLA (CM)'].min())
print('la estatura promedio de los nacidos en Sonora en 2023 es:', df_sonora['TALLA (CM)'].mean())
print()
print()
print("Cabe recalcar que no necesariamente los nacidos estaban vivos al momento de su concepción")
```
